```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

# for feature selection
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

# To visualize all the columns in the dataframe
pd.pandas.set_option('display.max_columns',None)
```

```python
dataset = pd.read_csv('X_train_house_price.csv')

# print shape of dataset with rows and columns
print(dataset.shape)
# print top 5 records
dataset.head()
```

sklearn error ValueError: Input contains NaN, infinity or a value too large for dtype('float64')

```python
np.any(np.isnan(dataset))
```

```python
np.all(np.isfinite(dataset))
```

```python
data_clean=dataset.copy()
def clean_dataset(data_clean):
    assert isinstance(data_clean, pd.DataFrame),"df needs to be a pd.DataFrame"
    data_clean.dropna(inplace=True)
    indices_to_keep = ~data_clean.isin([np.nan, np.inf, -np.inf]).any(1)
    return data_clean[indices_to_keep].astype(np.float64)
```

```python
clean_dataset(data_clean)
```

```python
np.any(np.isnan(data_clean))
```

```python
np.all(np.isfinite(data_clean))
```

### Seperating Dependent Variable from Dataset

```python
'Separate the sheep from the Goats'
```

```python
y_train=data_clean['SalePrice']
```

```python
x_train=data_clean.drop(['SalePrice'],axis=1)
```

### Applying Feature Selection
+ First lets specify the Lasso Regression model
+ Select a suitable alpha value(equivalent of panelty)
+ The bigger the alpha value the less features will be selected

```python
# We will use the SelectFromModel from sklearn which will select the features which has coefficients of non-zero
feature_sel_model=SelectFromModel(Lasso(alpha=0.005, random_state=0))
# remember the seed value, that is the random state in this function
feature_sel_model.fit(x_train, y_train)
```

```python
feature_sel_model.get_support()
```

### Let's print the number of total and selected features
+ This is how we can make a list of the selected features

```python
selected_feature=x_train.columns[(feature_sel_model.get_support())]

# Let's print some stats
print('Total Features: {}'.format((x_train.shape[1])))
print('Selected Features: {}'.format(len(selected_feature)))
print('Feature with coefficients shrank to zero {}'.format(np.sum(feature_sel_model.estimator_.coef_==0)))
```

```python
x_train=x_train[selected_feature]
```
